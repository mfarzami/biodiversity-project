#Introduction
---

Corals are among the oldest organisms still to be found on earth today first appearing in the Ordovician period, which began some 485 million years ago. This is remarkable compared to humans who have been around for only 7 million years and dinosaurs that lived 240 million years ago. Some corals are more than a million years old. Corals cover .1 percent of the ocean surface but host at least 25% of marine life. Because of this corals are known as the rainforest of the ocean because of its unparaellelled biodiversity, ie the amount of variance between species and their genetics. Biodiversity is critical for a stable ecosystem as it makes the food web more resilient to the changes from global warming, disease, and habitat fragmentation. Coral maintains a high level of species diversity compared to animals higher on the food chain. The Great Barrier Reef, for example, contains four hundred species of coral and only six species of turtles. For this reason, and because of weak data, we decided to change our projects focus from climate changes impact on sea turtles to corals as they are a better representation of sucessful vs failing biodiversity. In our dataset we were able to track seventeen types of corals whereas our previous dataset only tracked four species of turtle, unreliably and which few sightings too. Corals are immensly important to humans as well. 500 million people rely upon coral reefs for employment and food adding almost $7 trillion to the global economy each year. Moreover, coral reefs protect people living on the water from powerful storms. The ridges in coral reefs can reduce wave energy by up to 95%, creating a critical barrier to tsunamis and powerful storms that would otherwise inflict damage upon the shore. Furthermore, corals biodiversity makes it a key source of new medical breakthroughs. Coral extracts have been used to develop treatments for asthm, arthisis, cancer, and heartdisease. For example, corals amazing abiity to repair damaged DNA inflicted by sundamage is currently being studied to develop cancer treatment.  

In light of the amazing gift coral reefs give us it is distressing to learn 10 % of corals have been killed by climate change and a further 60% are threatened. Unlike other for a habitat destruction, coral reef collapse doesn't provide any direct benefits to humans the way deforestation does. In fact, coralreefs are far ore profitable alive as they provide essential fish grounds. So if humans aren’t actively targeting coral reef the question is what is killing them. Our answer to this question is carban emmision. We suspect that coral collapse is a direct result of increased CO2 and methane emmisions increasing global warming.

 
Corals are made up of thousands of animals called polyps. Polyps are related to sea anemones and jellyfish. They build their skeleton by depositing calcium carbonate, or limestone, used by other creatures to build their shells, onto a hard surface. This then forms a cup-like indentation where the stomach is stored and opens at one end. This stomach opening called its mouth, which it uses to eat and expel waste, is surrounded by a circle of tentacles used by the polyp to defend itself, capture small animals for food, and clear away debris. Because coral are rooted in one place this form of filter-feeding is unreliable. To compensate for this, corals form symbiotic relationships with algae, called zooxanthellae, which photosynthesize and 
share some of the byproducts, sugar, and oxygen, with the coral which in turn use these products to respire and give CO2 back to the algae. Corals receive up to ninety percent of their food from this symbiotic relationship and their distinctive coloring. When this relationship breaks down corals dispel all the algae and turn into white skeletons. These skeletons may recover given the correct conditions but if they are left out of balance the will starve and die. 
 
Conditions that can result in this breakdown are anything that dramatically changes the condition of the water causing the coral to panic. This includes changing temperature, acidity, agitation from hurricanes, and salinity. Global warming changes all these factors. Increased CO2 causes a build-up of gases that trap heat in the atmosphere increasing global temperature most of this excess heat is absorbed by the ocean but due to the ocean’s high specific heat and volume, this effect is limited to a change of a few degrees. This change in ocean temperature, while only a degree or two, is enough to stress the corals into expelling its algae in hopes better conditions will bring compatible algae in the future. Most of this excess CO2 in the atmosphere is also absorbed by the oceans through photosynthesis by phytoplankton and dissolving into the water, forming carbonic acid which increazes the acidity of the oceans. On top of breaking down this important sybiotic relationship, higher acidity also slows coral growth rates and weakens its skeletons. Global warning also results in a change in salintity because melting icecaps add more freshwater to the ocean reducing slainity and upsetting corals natural balance. Finally global warming has resulted in more frequent and severe storms that disrupt and damage corals. As a result, we will examine this relationship between increased CO2 and methane emmisions and their impact of global warming and coral collapse.

#Data Cleaning
---

```python
import pandas as pd
import numpy as np
from google.colab import drive
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from scipy.optimize import curve_fit
from matplotlib import rcParams
drive.mount("/content/gdrive")
animal_df = pd.read_csv('gdrive/My Drive/Data Science Project 1/animaldata.csv')
global_df = pd.read_csv('gdrive/My Drive/Data Science Project 1/globaltemp.csv')
methane_df = pd.read_csv('gdrive/My Drive/Data Science Project 1/methane_emissions.csv')
carbon_df = pd.read_csv('gdrive/My Drive/Data Science Project 1/carbon_emissions.csv')
corals_df = pd.read_csv('gdrive/My Drive/Data Science Project 1/deep_sea_corals.csv')
```

```python
#subsetting global temperature file to relevant columns and drop NaN
global_df = global_df.iloc[:,[0,7]]
global_df = global_df.dropna()

#renaming columns
global_df = global_df.rename(columns={'dt': 'date', 'LandAndOceanAverageTemperature': 'global_temp'})

#convert date column to date-time object format and remove day and month because we will only be analyzing the year
global_df['date'] = pd.to_datetime(global_df['date'], format='%Y-%m-%d').dt.strftime('%Y')

#averaging data points by year to find average temperature per year
unique_years = global_df['date'].unique()
data = {}
for year in unique_years:
  applicable_df = global_df[global_df['date'] == year]
  data[float(year)] = applicable_df['global_temp'].mean()

#making a dataset of the meaned values
finalglobal_df = pd.DataFrame(data.items(), columns = ['date','global_temp'])
finalglobal_df.head()
```

```python
#subetting methane emissions file
methane_df = methane_df.set_index('Country Name')
methane_df = methane_df.dropna(axis = 1, how='all')
methane_df = methane_df.loc[['World']]
methane_df = methane_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
methane_df = methane_df.rename(index= {'World':'world_methane_sum'})

#subetting carbon emissions file
carbon_df = carbon_df.set_index('Country Name')
carbon_df = carbon_df.dropna(axis = 1, how='all')
carbon_df = carbon_df.loc[['World']]
carbon_df = carbon_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
carbon_df = carbon_df.rename(index={'World':'world_carbon_sum'})

#merge carbon and methane datasets 
emissions_df = carbon_df.append(methane_df)
emissions_df = emissions_df.dropna(axis = 1, how='any')
emissions_df = emissions_df.transpose()
emissions_df = emissions_df.reset_index()
emissions_df = emissions_df.rename(columns={'index':'date'})
emissions_df.head()
```

```python
#subsetting corals dataframe 
corals_df['ObservationDate'] = pd.to_datetime(corals_df.ObservationDate).dt.strftime('%Y')
corals_df = corals_df[['ScientificName', 'VernacularNameCategory', 'ObservationDate', 'DepthInMeters', 'Locality']]
corals_df = corals_df.rename(columns={'ScientificName':'scientific_name', 'VernacularNameCategory': 'vernacular_name', 'ObservationDate':'date', 'DepthInMeters':'depth_m', 'Locality':'locality'})
corals_df = corals_df.dropna(axis = 1, how='all')
corals_df = corals_df.dropna()
corals_df.head()
```

#Data Analysis
---

We decided to start our data analysis by tracking the global trends of methane and carbon emissions over time, and calculating the strength of the trends in order to later see whether emissions could be a causal factor in our other variable trends.

```python
#Create plot
plt.figure(figsize=(10,7))
plt.plot(emissions_df['date'], emissions_df['world_carbon_sum'], c='blue', label = 'Carbon Emissions')
plt.plot(emissions_df['date'], emissions_df['world_methane_sum'], c='orange', label = 'Methane Emissions')
plt.legend()
plt.xlabel('Year (1970-2015')
plt.ylabel('Emissions (kt)')
plt.locator_params(axis="x", nbins=4)
plt.show()

#Find coefficient and r-squared for carbon emissions
lr = LinearRegression().fit(emissions_df['date'].values.reshape(-1, 1), emissions_df['world_carbon_sum'].values.reshape(-1, 1))
coef = lr.coef_
regvalue = lr.score(emissions_df[['date']], emissions_df['world_carbon_sum'])
print('The Coefficient for carbon emissions is: ' + str(lr.coef_[0][0].round(2)))
print('The R-Squared value for carbon emissions is: ' + str(round(regvalue,2)))

#Find coefficient and r-squared for methane emissions
lr1 = LinearRegression().fit(emissions_df['date'].values.reshape(-1, 1), emissions_df['world_methane_sum'].values.reshape(-1, 1))
coef1 = lr1.coef_
regvalue1 = lr1.score(emissions_df[['date']], emissions_df['world_methane_sum'])
print('The Coefficient for methane emissions is: ' + str(lr1.coef_[0][0].round(2)))
print('The R-Squared value for methane emissions is: ' + str(round(regvalue1,2)))
```

Over the past century, global greenhouse gas emission rates have been increasing steadily, as indicated by the graph above. Greenhouse gas emissions are impacted by a wide variety of natural and human activity, but given the great industrial advancements made to further human society, environmental issues such as increasing ocean levels (due to melting ice regions) and marine organism degredation have become more prevalent. In the scope of our project, coral reefs are suffering as marine heat waves begin to become overwhelming in temperature, ultimately causing coral bleaching and death. The R^2 value for both plots are close to 1, making us confident in the correlation between emissions and years. Although both gases are increasing in terms of emissions, we noticed that overall, carbon is much more prevalant, which we found in our research is the most commonly produced greenhouse gas, especially from factories.

Next, we focused on our temperature data to see if there was a similar strong positive correlation between average global temperature and years.

```python
#graph the change in global temperature
plt.figure(figsize=(10,7))
plt.scatter(finalglobal_df['date'], finalglobal_df['global_temp'], alpha=0.5)
plt.xlabel('Years')
plt.ylabel('Average Global Temperature (c)')

#find the strength of correlation and the regression line
model1 = LinearRegression().fit(finalglobal_df['date'].values.reshape(-1, 1), finalglobal_df['global_temp'].values.reshape(-1, 1))
regvalue1 = model1.score(finalglobal_df['date'].values.reshape(-1, 1), finalglobal_df['global_temp'].values.reshape(-1, 1))
plt.plot(finalglobal_df['date'].values.reshape(-1, 1), model1.coef_*np.array(finalglobal_df['date'].values.reshape(-1, 1))+model1.intercept_, c='orange')
plt.show()
print('The Coefficient for carbon emissions is: ' + str(model1.coef_[0][0].round(2)))
print('The R-Squared value for carbon emissions is: ' + str(round(regvalue1,2)))
print('The intercept is: ' + str(round(model1.intercept_[0],2)))
```

Although there are fluctuations, this graph demonstrates the general increase in average global temperature over the past two centuries. According to our research, if the average global warming reaches 2 degrees celsius higher than what was measured in 1970, more than 70% of Earth's coastlines will see sea-level rise greater than 0.2 meters, resulting in drastic coastal flooding, beach erosion, salinization of water supplies and other impacts on humans and ecological systems that include heavy economic, social and environmental damages. We noticed that the R^2 was pretty high, but visually we also saw a curvature, as the values were increasing seemingly exponentially as time went on. This realization allowed us to try an exponential model to see if it would be more accurate before we attempted prediction.

```python
#Find exponential regression line for years vs. average global temperature 
def quadratic(X, a, b, c): 
    return a*X**2 + b*X + c
plt.figure(figsize=(10,7))
out1, out2 = curve_fit(quadratic, finalglobal_df['date'], finalglobal_df['global_temp'])
plt.scatter(finalglobal_df['date'], finalglobal_df['global_temp'], alpha=0.5)
predict_line = plt.plot(finalglobal_df['date'].sort_values(), quadratic(finalglobal_df['date'].sort_values(), *out1), c='orange')
plt.xlabel('Years')
plt.ylabel('Average Global Temperature (c)')
plt.tight_layout()
print(f"Exponential model r-squared: {np.corrcoef(quadratic(finalglobal_df['date'], *out1), finalglobal_df['global_temp'])[0,1]**2:.3f}")

#Extend regression line to 2070 to predict future
plt.xlim(1850, 2070)
x_extended = np.linspace(2015,2070)
predict_line = plt.plot(x_extended, quadratic(x_extended, *out1), c='orange')
plt.show()

#Find useful predictions based on model 
x,y = predict_line[0].get_data()
predict_temp = np.interp(2031, x,y)
print('In 2031, we predict the global average temperature will be '+str(predict_temp.round(2))+' degrees celsius')
predict_year = np.interp(16.5, y, x)
print('We predict the global average temperature will reach a 1.5 degree celsius increase by the year '+str(predict_year.round(2)))
```

We also decided to use an exponential model (quadratic function) to track the increase in global temperature because we saw that average global temperature is rising at an increasingly faster rate so the linear function is not necessarily the most representative model. We found that the model is much more aligned to the data this way, as seen by the increased r^2 value.

We used our exponential model to find the year in which the average global temperature will have increased 1.5 degrees celsius from preindustrial times (we considered this to mean from 1850 on). We found that it is predicted to reach this increase by around 2058. The increase of 1.5 degrees is impactful because many scientists believe it will mark a negative turning point in climate change. The Intergovernmental Panel on Climate Change (IPCC) predicts that this turning point will cause heatwaves, ocean levels rising, and the destruction of 70-90% of coral reefs. 

We also used our prediction model to find what we think the global average temperature will be in ten years from now. We will use this temperature of 16.08 degrees celsius in a multivariable linear regression prediction of coral presence later on.

While the R² value was very low, we did see a slight negative correlation between methane emissions and coral sightings, which could be a result of negatively impacted ocean health due to greenhouse gases. We aimed to test the same idea with carbon emissions next.

```python
#graph carbon emissions effect on coral sightings
plt.figure(figsize=(10,7))
plt.scatter(recentcorals_merged['world_carbon_sum'], recentcorals_merged['sightings'])
plt.xlabel('Carbon Emissions')
plt.ylabel('Coral Sightings')

#find the regression line
model4 = LinearRegression().fit(recentcorals_merged['world_carbon_sum'].values.reshape(-1, 1), recentcorals_merged['sightings'].values.reshape(-1, 1))
regvalue4 = model4.score(recentcorals_merged['world_carbon_sum'].values.reshape(-1, 1), recentcorals_merged['sightings'].values.reshape(-1, 1))
plt.plot(recentcorals_merged['world_carbon_sum'].values.reshape(-1, 1), model4.coef_*np.array(recentcorals_merged['world_carbon_sum'].values.reshape(-1, 1))+model4.intercept_, c='orange')
print('The Coefficient is: ' + str(model4.coef_[0][0].round(5)))
print('The R-Squared value is: ' + str(round(regvalue4,2)))
print('The intercept is: ' + str(round(model4.intercept_[0],2)))
plt.show()
```

Here, we see that carbon emmisions and methane emmisions have an r squared of around .4 which is notable but not very strong. This is consistent with our intuition because most of this excess CO2 in the atmosphere is absorbed by the oceans through photosynthesis by phytoplankton and dissolving into the water, forming carbonic acid which increazes the acidity of the oceans. On top of breaking down the important sybiotic relationship between corals and algae, called zooxanthellae, higher acidity also slows coral growth rates and weakens its skeletons increasing the likelyhood of mortality.

Next, we will train a multivariable linear regression model that uses the inputs global_temp, world_carbon_sum, and world_methane_sum jointly to predict coral sightings. We hope to see a more effective model from combining all the input variables together.

```python
multi_var = LinearRegression().fit(recentcorals_merged[['global_temp', 'world_carbon_sum', 'world_methane_sum']], recentcorals_merged['sightings'])
print('The Coefficient for global_temp is: ' + str(multi_var.coef_[0].round(2)))
print('The Coefficient for world_carbon_sum is: ' + str(multi_var.coef_[1].round(5)))
print('The Coefficient for world_methane_sum is: ' + str(multi_var.coef_[2].round(5)))
print('The Intercept is: ' + str(multi_var.intercept_.round(2)))
print('The R-squared score of the model is: ' + str(multi_var.score(recentcorals_merged[['global_temp', 'world_carbon_sum', 'world_methane_sum']], recentcorals_merged['sightings'])))
```

Here we see that the multivariable linear regression created a higher R² value than any other that we saw in regards to coral sightings. Although it is still not very high, we do think that it shows that the variables work together to cause coral sightings to decrease.
